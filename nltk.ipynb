```python
import nltk
import json
from nltk.stem.porter import *
from nltk.text import Text
from nltk.probability import ConditionalFreqDist
from nltk.stem import WordNetLemmatizer
import heapq
from gensim import corpora, models, similarities
nltk.download('punkt')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
json_str = posDict
dataset = json.loads(json_str)
nltk.download('punkt')
dataset['post'][0]['text']
dataset['post'][0]['timestamp']
dataset['post'][0]['username']
```

```python
import time
jsondict = dict()  
dataCollectWord = dict()
termDf = dict()
term_set= set() #所有相異詞
item = 0
cnt = -1
monthlist = []
monthdict=dict()
for data in dataset['post'] : 
  struct_time = time.localtime(data['timestamp']) 
  timeString = time.strftime("%Y-%m", struct_time) # 轉成字串
  if timeString in monthlist:
    monthdict[cnt]['datacnt']+= 1# 有幾個文章?
    monthdict[cnt]['dataid'].append(item)
  else :
    cnt+=1
    monthlist.append(timeString)
    monthdict[cnt]=dict()
    monthdict[cnt]=dict()
    monthdict[cnt]['datacnt'] = 1
    monthdict[cnt]['df']=dict()
    monthdict[cnt]['dataid']=list()
    monthdict[cnt]['dataid'].append(item)
  paragraph_str=data['text']
  token_set = paragraph_str.split()
  lower_set=set()
  term_set = set()
  dataCollectWord[item] = dict()
  for token in token_set:
    lowertoken = token.lower()
    lower_set.add(lowertoken)
    term_set.add(lowertoken)
    if lowertoken in dataCollectWord[item] : 
      dataCollectWord[item][lowertoken]+=1
    else : 
      dataCollectWord[item][lowertoken] = 1
    if lowertoken in termDf :
       termDf[lowertoken]+=1
    else : 
       termDf[lowertoken] = 1
    if "#"  not in token :
      if lowertoken  in monthdict[cnt]['df']:
        monthdict[cnt]['df'][lowertoken]+=1
      else:
       monthdict[cnt]['df'][lowertoken] = 1
  jsondict[item]=lower_set
  item+=1
```

```python
dataset = json.loads(posDict)
token_set = []
base_data = []
for data in dataset['post']:
    base_data.append(data['text'])
base_items = [[i for i in nltk.word_tokenize(item)] for item in base_data]
dictionary = corpora.Dictionary(base_items)
corpus = [dictionary.doc2bow(item) for item in base_items]
tf = models.TfidfModel(corpus)
num_features = len(dictionary.token2id.keys())
index = similarities.MatrixSimilarity(
    tf[corpus], num_features=num_features)

test_text = input("")
test_words = [word for word in nltk.word_tokenize(test_text)]
print(test_words) 
new_vec = dictionary.doc2bow(test_words)
sim = index[tf[new_vec]]
sim = sim.tolist()
max_num_index_list = map(sim.index, heapq.nlargest(3, sim))
print("Search:"+test_text)
count = 0
for max_index in max_num_index_list:
    if max_index!=0:
      count += 1
      print("Paste"+str(count)+":"+base_data[max_index])
if count==0:
  print("沒有找到，非常抱歉")
```
