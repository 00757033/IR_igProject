```python
# postSum= 0
def getData(account): 
  url= 'https://www.instagram.com/'+ account+ '/?__a=1'
  cookies= {'sessionid': '8340972345%3AzYYS3Fon00LsSK%3A3'} # 複製cookie，修改sessionid 
  res_data= requests.get(url, cookies= cookies).content
  data= json.loads(res_data)

  print('\n-',account)
  userData= data['graphql']['user']
  postsCount= userData['edge_owner_to_timeline_media']['count'] # 總貼文數

  user= dict()
  user['username']= userData['username'] # 使用者帳號
  user['biography']= userData['biography'] # 自我介紹
  user['followedCount']= userData['edge_followed_by']['count'] # 粉絲人數
  userid= userData['id']
  users.append(user)
  # postSum+= postsCount
  print('有', postsCount, '則')

  check= True
  endCursor= ''
  nowww= 12
  media= data['graphql']['user']['edge_owner_to_timeline_media']
  first= 50

  while check:
    if endCursor!= '': # 不是第一頁貼文，要重新request
      url= 'https://www.instagram.com/graphql/query/?query_id=17888483320059182&id='+ str(userid)+ '&first='+ str(first)+'&after='+ endCursor
      # first是希望返回的照片數量
      # cookies= {'sessionid': '1560442426%3A1SvcFb6x3RoL0f%3A27'}
      res_data= requests.get(url, cookies= cookies).content
      data= json.loads(res_data)
      if data['status']!= 'ok': # fail
        print(data['status'])
        break
      nowww+= first
      media= data['data']['user']['edge_owner_to_timeline_media']
    
    edges= media['edges'] # 一頁12篇貼文
    for edge in edges:
      if not edge['node']['edge_media_to_caption']['edges'] :
        print('edge is Null')
        continue

      # message= edge['node']['edge_media_to_comment']['count'] # 圖片貼文留言數
      post= dict()
      post['username']= userData['username']
      # post['likeCount']= edge['node']['edge_liked_by']['count'] # 貼文按讚數
      post['text']= edge['node']['edge_media_to_caption']['edges'][0]['node']['text'] # 貼文內容
      post['timestamp']= edge['node']['taken_at_timestamp'] # 貼文時間
      posts.append(post)
      
    
    print('(', nowww, ',', len(posts), ')', end= ', ')
    if nowww> 1000:
      break

    # 是否有下一頁文章
    page_info= media['page_info']
    check= page_info['has_next_page']
    
    if check == True:
      endCursor= page_info['end_cursor']
```

```python
import time

jsondict = dict()  
dataCollectWord = dict()
termDf = dict()
term_set= set() #所有相異詞
item = 0
cnt = -1
monthlist = []
monthdict=dict()

for data in dataset['post'] : 
  struct_time = time.localtime(data['timestamp']) 
  timeString = time.strftime("%Y-%m", struct_time) # 轉成字串
  if timeString in monthlist:
    monthdict[cnt]['datacnt']+= 1# 有幾個文章?
    monthdict[cnt]['dataid'].append(item)
  else :
    cnt+=1
    monthlist.append(timeString)
    monthdict[cnt]=dict()
    monthdict[cnt]=dict()
    monthdict[cnt]['datacnt'] = 1
    monthdict[cnt]['df']=dict()
    monthdict[cnt]['dataid']=list()
    monthdict[cnt]['dataid'].append(item)
    monthdict[cnt]['data']=timeString
    
    
  paragraph_str=data['text']
  token_set = paragraph_str.split()
  lower_set=set()
  term_set = set()
  dataCollectWord[item] = dict()
  for token in token_set:
    
    lowertoken = token.lower()
    
    if lowertoken!='newsweek':
      lower_set.add(lowertoken)
      term_set.add(lowertoken)
    if lowertoken in dataCollectWord[item] : 
      dataCollectWord[item][lowertoken]+=1
    else : 
      dataCollectWord[item][lowertoken] = 1
    if lowertoken in termDf :
       termDf[lowertoken]+=1
    else : 
       termDf[lowertoken] = 1
    if "#" in token :
      if lowertoken[lowertoken.find('#')+1:]  in monthdict[cnt]['df']:
        monthdict[cnt]['df'][lowertoken[lowertoken.find('#')+1:]]+=1
      else:
       monthdict[cnt]['df'][lowertoken[lowertoken.find('#')+1:]] = 1
  jsondict[item]=lower_set
  item+=1
```

```python
dataset = json.loads(posDict)
token_set = []
base_data = []
for data in dataset['post']:
    base_data.append(data['text'])
base_items = [[i for i in nltk.word_tokenize(item)] for item in base_data]
dictionary = corpora.Dictionary(base_items)
corpus = [dictionary.doc2bow(item) for item in base_items]
tf = models.TfidfModel(corpus)
num_features = len(dictionary.token2id.keys())
index = similarities.MatrixSimilarity(
    tf[corpus], num_features=num_features)

test_text = input("")
test_words = [word for word in nltk.word_tokenize(test_text)]
print(test_words) 
new_vec = dictionary.doc2bow(test_words)
#print(new_vec)
sim = index[tf[new_vec]]
sim = sim.tolist()
#print(len(sim))
max_num_index_list = map(sim.index, heapq.nlargest(20, sim))
print("Search:"+test_text)
count = 0
max_num_index_list=sorted(max_num_index_list, reverse = True)
for max_index in max_num_index_list:
    #print(max_index)
    if max_index>200:
      count += 1
      print("Paste"+str(count)+":"+base_data[max_index])
if count==0:
  print("沒有找到，非常抱歉")
print(count)
```
